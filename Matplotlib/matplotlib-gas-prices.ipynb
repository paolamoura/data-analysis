```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Gas Prices/gas_prices.csv')
gas
```

```python
# Criando o primeiro gr√°fico.

plt.plot(gas['Year'], gas['Australia'])

plt.show()
```

```python
plt.plot(gas['Year'], gas['Australia'])
plt.plot(gas['Year'], gas['Italy'])
plt.plot(gas['Year'], gas['USA'])

plt.show()
```

```python
# Adicionando Legendas e Ticks.

plt.title('Valor do Combust√≠vel em US$')

plt.plot(gas['Year'], gas['Australia'], label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], label = 'Italy')
plt.plot(gas['Year'], gas['USA'], label = 'USA')
plt.plot(gas['Year'], gas['Japan'], label = 'Japan')

plt.xticks(gas['Year'][::2]) # Mostra anos pulando de dois em dois.

plt.legend()
plt.show()
```

```python
# Formatando label e cores.

plt.title('Valor do Combust√≠vel em US$')

plt.plot(gas['Year'], gas['Australia'], 'r.-', label = 'Australia') #'r.-' √© cor red + delimitador.
plt.plot(gas['Year'], gas['Italy'], 'y.-', label = 'Italy') #'r.-' √© cor yellow + delimitador.
plt.plot(gas['Year'], gas['USA'], 'b.-', label = 'USA') #'r.-' √© cor blue + delimitador.
plt.plot(gas['Year'], gas['Japan'], 'g.-', label = 'Japan') #'r.-' √© cor green + delimitador.

plt.xticks(gas['Year'][::2]) # Mostra anos pulando de dois em dois.

plt.xlabel('Years') # Adiciona t√≠tulo no eixo x.
plt.ylabel('US$') # Adiciona t√≠tulo no eixo y.


plt.legend()
plt.show()
```

```python
# Adicionando com o for loop.

plt.title('Price is US$')

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = [country], marker = '.')

plt.xticks(gas['Year'][::2]) # Mostra anos pulando de dois em dois.

plt.xlabel('Years') # Adiciona t√≠tulo no eixo x.
plt.ylabel('US$') # Adiciona t√≠tulo no eixo y.


plt.legend()
plt.show()
```

```python
# Salvando gr√°fico no Matplotlib.

plt.title('Price is US$')

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = [country], marker = '.')

plt.xticks(gas['Year'][::2]) # Mostra anos pulando de dois em dois.

plt.xlabel('Years') # Adiciona t√≠tulo no eixo x.
plt.ylabel('US$') # Adiciona t√≠tulo no eixo y.

plt.legend()
plt.savefig('country.png')
plt.savefig('country.pdf')
plt.show()
```
