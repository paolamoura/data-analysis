```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ice Cream/ice-cream.csv')

df
```

```python
# Mostra valores de temperatura do maior para o menor.

result = df.sort_values('Temperature (Celsius)', ascending = False)

result
```

```python
# Construindo gráfico.

plt.plot(result['Temperature (Celsius)'], result['Ice Cream Sales'], 'r.-', label = 'US$')
plt.title('Ice Cream Truck')
plt.ylabel('Ice Cream Sales (US$)')
plt.xlabel('Temperature (Celsius)')
plt.legend()

plt.savefig('icecream-sales.png')
plt.savefig('icecream-sales.pdf')

plt.show()
```
