```python
import pandas as pd
import re
```

```python
#DataFrame é a junção de colunas e linhas.

df = pd.read_csv('Melbourne House Prices Less/house.csv')
```

```python
#Filtrando dados usando condições lógicas.

#Exemplo 1: Filtrando número de casas com 3 quartos e preço menor ou igual a quinhentos mil.

filter1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] <= 500000)]
filter1
```

```python
#Filtrando dados usando expressões regulares para considerar letras minúsculas e maiúsculas.

#Exemplo 2: Filtrando casas de uma rua (street) ou estrada (road) chamada Turner.

filter2 = df.loc[df['Address'].str.contains('turner st|Turner Rd', flags = re.I)]
filter2
```

```python
#Filtrando com RegX e condições.

#Exemplo 3: Filtrando casas com número 59 e com preço menor ou igual a quinhentos mil.

filter3 = df.loc[df['Address'].str.contains('^59', flags = re.I) & (df['Price'] <= 500000)]
filter3
```

```python
#Modificando dados na DataFrame.

#Exemplo 4: Troca o nome do vendedor Nelson para Ronald.

df.loc[df['SellerG'] == 'Nelson', 'SellerG'] = 'Ronald'
df
```

```python
#Exemplo 4: Nas linhas que tenham Ronald como vendedor, muda os dados das colunas Method e Type para "Pending".

df.loc[df['SellerG'] == 'Ronald', ['Method', 'Type']] = 'Pending'
df
```

```python
#Agrupando dados com o Groupby. Agrupa pela média, soma, etc.

#Exemplo 5: Agrupa os dados por vendedores e passa a função soma que lista de valores baseado no preço. A listagem é feita da pessoa que mais vende, para a que menos vende. O head irá mostrar somente os 10 melhores vendedores.

df.groupby(['SellerG']).sum().sort_values('Rooms', ascending = False).head(10)
```
